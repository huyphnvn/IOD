````python
# sample points 
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    '''
    Return m, b
    # ANSWER
    '''
    #Covert list to array
    X = np.array(X)
    Y = np.array(Y)
    
    m = (len(X) * np.dot(X, Y) - np.sum(X) * np.sum(Y)) / (len(X) * np.sum(X**2) - np.sum(X)**2)
    b = (np.sum(Y) - m * np.sum(X)) / len(X)
    
    return m, b

# solution
m, b = best_fit(X, Y)
print(m, b)
#best fit line:
#y = 0.80 + 0.92x

# plot points and fit line
# ANSWER
sns.scatterplot(x=X, y=Y)
sns.lineplot(x=X, y=0.8 + 0.92*np.array(X), color='orange')
````

````python
#def mean_confidence_interval(data, confidence=0.95):
# Calculate m, h
# ANSWER
#m, se = ???, ???
#h = ???
#return m, m-h, m+h

def mean_confidence_interval(data, confidence=0.95):
  a = 1.0 * np.array(data) #Why do we have to multiply the array with 1?
  n = len(a)
  m, se = np.mean(a), scipy.stats.sem(a)
  h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
  return m, m-h, m+h
data = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
data = [170,167,	174,	179,	179,	187,	179,	183,	179 ,156,	163	,156	,187,	156	,
        167	,156	,174	,170 ,183	,179	,174	,179	,170,	159	,187	]
print(mean_confidence_interval(data))

results = stats.ttest_1samp(data,170,alternative='greater')
````
